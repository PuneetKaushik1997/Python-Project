```python
# Group by 'month' and response
df.groupby('Month_no.')['response'].mean().plot(kind='line')
plt.title('Response by Month')
plt.show()

# If 'day' is related to specific days of the week, you can explore:
df.groupby('day')['response'].mean().plot(kind='line')
plt.title('Response by Day of the Month')
plt.show()
```

# Response by day of month
The response rate is highest at the beginning of the month, shows unstable variations in the middle, and experiences mixed patterns toward the end, including both lows and sporadic increases.

## 9. Outlier Detection and Handling

```python
from scipy import stats


# Calculate Z-scores for numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])
z_scores = stats.zscore(numeric_df)

# Convert Z-scores to a DataFrame for better visibility
z_scores_df = pd.DataFrame(z_scores, columns=numeric_df.columns)

# Display the first few rows of Z-scores
print(z_scores_df.head())
```

```python
# Identify rows where any Z-score is greater than 3 (outliers)
outliers = numeric_df[(z_scores_df > 3).any(axis=1)]

# Display the outliers
print(outliers)
```
